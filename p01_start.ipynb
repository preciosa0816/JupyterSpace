# p01.Numpy,Pandas, Matplotlib 를 활용한 데이터 분석 실습

```python
#그래프를 바로 보기 위한 명령어 

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
```

```python
# 50개의 난수로 이루어진 데이터 생성 
data=np.random.rand(50)
print(data)
```

```python
#위에서 발생한 데이터를 pandas Series형 타입으로 데이터 변환
seri=pd.Series(data)
print(seri)
```

```python
#배열에 들어가 있는 데이터를 선형 그래프로 그리기 
seri.plot()
```

```python
#그래프 스타일 변경하기
matplotlib.style.use('ggplot')
seri.plot()
```
